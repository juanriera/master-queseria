```python
# Extraer recetas únicas
recetas_unicas = df_fab_queso['receta'].dropna().drop_duplicates().tolist()

# Calcular número de dígitos necesarios (por ejemplo, 2 para hasta 99 recetas)
n_digitos = len(str(len(recetas_unicas)))

# Crear diccionario con ceros a la izquierda
diccionario_recetas = {
    receta: f"receta {str(i+1).zfill(n_digitos)}"
    for i, receta in enumerate(recetas_unicas)
}

# Aplicar el diccionario al DataFrame
df_fab_queso['receta'] = df_fab_queso['receta'].map(diccionario_recetas)
```

```python

tabla_cruzada = pd.crosstab(df_fab_queso['receta'], df_fab_queso['formato'])

# Reemplazar ceros por cadenas vacías
tabla_cruzada = tabla_cruzada.replace(0, "")

# Mostrar la tabla
print("📊 Tabla cruzada de recuentos receta/producto (sin ceros):")
display(tabla_cruzada)
```

```python
# Diccionario de recodificación
formato_normalizado = {
    '1 kg': '1 kg',
    '3 kg': '3 kg',
    '3kg': '3 kg',
    'barra 3 kg': 'barra 3 kg',
    'barra 3kg': 'barra 3 kg',
    'barra 4,2': 'barra 4,2 kg',
    'barra 4.2': 'barra 4,2 kg',
    'barra.4.2': 'barra 4,2 kg',
    'mini': 'mini',
    'minis': 'mini'
}
# Aplicar la recodificación
df_fab_queso['formato'] = df_fab_queso['formato'].map(formato_normalizado)
```

```python

tabla_cruzada = pd.crosstab(df_fab_queso['receta'], df_fab_queso['formato'])

# Reemplazar ceros por cadenas vacías
tabla_cruzada = tabla_cruzada.replace(0, "")

# Mostrar la tabla
print("📊 Tabla cruzada de recuentos receta/producto (sin ceros):")
display(tabla_cruzada)
```
